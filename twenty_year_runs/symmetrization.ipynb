```python
# Loading libraries

import xarray as xr
import matplotlib.pyplot as plt
import glob
import numpy as np
```

# Testing Symmetrization

```python
path = glob.glob('/data/keeling/a/pappup2/a/CESM/cam_15_n/CESM_output_data/atm/hist/*')
single = path[239]
data_single = xr.open_dataset(single)
data_single
```

```python
control_path = glob.glob('/data/keeling/a/pappup2/a/CESM/cam5_new/CESM_output_data/atm/hist/*')
control_single = control_path[239]
control_single = xr.open_dataset(control_single)
control_single
```

```python
data_single.TS.mean('lon').plot()
plt.title('Northern Hemisphere 15 patch SST')
```

```python
# https://stackoverflow.com/questions/53121089/regridding-coordinates-with-python-xarray
south_h = data_single.assign_coords(lat=data_single.lat[::-1])
south_h
```

```python
south_h.TS.mean('lon').plot()
plt.title('Flipped NH 15 degree Patch SST')
```

```python
combine = data_single.TS.mean('lon') + south_h.TS.mean('lon')
combine_dSST = combine - 2*control_single.TS.mean('lon')
combine_dSST.plot()
plt.title('Flipped and Original SST Patches Together')
```

# Load Data

```python
#Control

data_control_path = glob.glob("/data/keeling/a/pappup2/a/CESM/cam5_new/CESM_output_data/atm/hist/*") #End of year control
data_control_last18=data_control_path[24:240] # Get years 3-20
data_control_data = xr.open_mfdataset(data_control_last18,combine="by_coords",use_cftime=True)
data_control = data_control_data.mean('time') # Average over the last 18 years
```

```python
#Symmetric data

data_15_both=glob.glob("/data/keeling/a/pappup2/a/CESM/cam_15/CESM_output_data/atm/hist/*")
data_15_last18_both=data_15_both[24:240] # Get years 3-20
data_15_data_both = xr.open_mfdataset(data_15_last18_both,combine="by_coords",use_cftime=True)
data_15 = data_15_data_both.mean('time') # Average over the last 18 years
```

```python
# Southern Hemisphere 15 degree patch

data_15_path_s=glob.glob("/data/keeling/a/pappup2/a/CESM/cam_15_s/CESM_output_data/atm/hist/*")
data_15_last18_s=data_15_path_s[24:240] # Get years 3-20
data_15_data_s = xr.open_mfdataset(data_15_last18_s,combine="by_coords",use_cftime=True)
data_15_s = data_15_data_s.mean('time') # Average over the last 18 years
```

```python
# Northern Hemisphere 15 degree patch

data_15_path=glob.glob("/data/keeling/a/pappup2/a/CESM/cam_15_n/CESM_output_data/atm/hist/*")
data_15_last18=data_15_path[24:240] # Get years 3-20
data_15_data = xr.open_mfdataset(data_15_last18,combine="by_coords",use_cftime=True)
data_15_n = data_15_data.mean('time') # Average over the last 18 years
```

# SST

```python
dSST_both = data_15.TS.mean('lon') - data_control.TS.mean('lon')
plt.rcParams['figure.figsize'] = [12,5]

plt.suptitle('Both Hemisphere Experiment')
plt.subplot(1,2,1)
data_15.TS.mean('lon').plot()
plt.title('SST')
plt.subplot(1,2,2)
dSST_both.plot()
plt.title('dSST')
plt.ylabel('dSST')
```

```python
dSST_north = data_15_n.TS.mean('lon') - data_control.TS.mean('lon')
plt.rcParams['figure.figsize'] = [12,5]

plt.suptitle('North Hemisphere Experiment')
plt.subplot(1,2,1)
data_15_n.TS.mean('lon').plot()
plt.title('SST')
plt.subplot(1,2,2)
dSST_north.plot()
plt.title('dSST')
plt.ylabel('dSST')
```

```python
dSST_south = data_15_s.TS.mean('lon') - data_control.TS.mean('lon')
plt.rcParams['figure.figsize'] = [12,5]

plt.suptitle('South Hemisphere Experiment')
plt.subplot(1,2,1)
data_15_s.TS.mean('lon').plot()
plt.title('SST')
plt.subplot(1,2,2)
dSST_south.plot()
plt.title('dSST')
plt.ylabel('dSST')
```

```python
# https://stackoverflow.com/questions/53121089/regridding-coordinates-with-python-xarray

dSST_flipped = data_15_n.TS.mean('lon').assign_coords(lat=list(reversed(data_15_n.lat))) - data_control.TS.mean('lon')
plt.rcParams['figure.figsize'] = [12,5]

plt.suptitle('Flipped North Hemisphere Experiment')
plt.subplot(1,2,1)
data_15_n.TS.mean('lon').assign_coords(lat=list(reversed(data_15_n.lat))).plot()
plt.title('SST')
plt.subplot(1,2,2)
dSST_flipped.plot()
plt.title('dSST')
plt.ylabel('dSST')
```

# Radiation

```python
SW_control = data_control.FSNT.mean('lon')
SW_control.load()

SW_both = data_15.FSNT.mean('lon') - SW_control
SW_both.load()
SW_n = data_15_n.FSNT.mean('lon') - SW_control
SW_n.load()
SW_s = data_15_s.FSNT.mean('lon') - SW_control
SW_s.load()
SW_flipped = SW_n.assign_coords(lat=list(reversed(data_15_n.lat)))
SW_flipped.load()

SW_ns = SW_n + SW_s
SW_nflip = SW_n + SW_flipped
```

```python
plt.subplot(1,2,1)
data_15_s.FSNT.mean('lon').plot()
plt.title('SH Experiment')
plt.subplot(1,2,2)
data_15_n.FSNT.assign_coords(lat=list(reversed(data_15_n.lat))).mean('lon').plot()
plt.title('Flipped NH Experiment')
```

```python
data_15_s.FSNT.mean('lon').plot(label='SH Experiment')
data_15_n.FSNT.assign_coords(lat=list(reversed(data_15_n.lat))).mean('lon').plot(label='Flipped NH')
plt.legend()
plt.title('FSNT')
```

```python
data_15_s.FSNT.mean('lon').plot(label='SH Experiment')
data_15_n.FSNT.mean('lon').plot(label='NH Experiment')
plt.title('SW TOA Radiation')
plt.legend()
```

```python
LW_control = data_control.FLNT.mean('lon')
LW_control.load()

LW_both = data_15.FLNT.mean('lon') - LW_control
LW_both.load()
LW_n = data_15_n.FLNT.mean('lon') - LW_control
LW_n.load()
LW_s = data_15_s.FLNT.mean('lon') - LW_control
LW_s.load()
LW_flipped = LW_n.assign_coords(lat=list(reversed(data_15_n.lat)))
LW_flipped.load()

LW_ns = LW_n + LW_s
LW_nflip = LW_n + LW_flipped
```

```python
plt.rcParams['figure.figsize'] = [8,5]

LW_both.plot(label='Both H Exp.')
LW_ns.plot(label='North, South Exp')
LW_nflip.plot(label='North,Flipped North')
plt.legend()
plt.title('LW TOA')
```

```python
TOA_both = SW_both - LW_both
TOA_n = SW_n - LW_n
TOA_s = SW_s - LW_s
TOA_flipped = SW_flipped - LW_flipped

TOA_ns = TOA_n + TOA_s
TOA_nflip = TOA_n + TOA_flipped
```

```python
plt.rcParams['figure.figsize'] = [8,5]

TOA_both.plot(label='Both H Exp.')
TOA_ns.plot(label='North, South Exp')
TOA_nflip.plot(label='North,Flipped North')
plt.legend()
plt.title('TOA Radiation')
plt.ylabel('Total Radiation')
```
