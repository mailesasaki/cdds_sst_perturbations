```python
def plotting(array,num,vmin,vmax,cmap,lat):
    """
    For plotting various 3-D variables
    """
    plt.subplot(gs[0,num])
    mappable = array.plot(vmin=vmin, vmax=vmax, yincrease=False, cmap=cmap, add_colorbar=False)
    plt.hlines(y=1000, xmin=lat-10, xmax=lat+10, colors='black', lw=5)
    plt.xticks([-90,-45,0,45,90])
    ax = pylab.gca()
    ax.set_xlabel('')
    plt.xticks(rotation=90)
    plt.title(lat)
    return mappable
```

```python
nrow = 1
ncol = 14
matplotlib.rcParams['figure.figsize'] = [50, 20]
fig = plt.figure(figsize=(20,45))
grid = ImageGrid(fig, 
                 111, # as in plt.subplot(111)
                 nrows_ncols=(nrow,ncol),
                 axes_pad=0,
                 share_all=True,)


im = np.random.rand(28,28)
gs = gridspec.GridSpec(14, 14)
gs.update(wspace=0.1)

vmin=-25
vmax=25
cmap='RdBu_r'

plotting(hum_00,0,vmin,vmax,cmap,0)
ax = pylab.gca()

plotting(hum_05,1,vmin,vmax,cmap,5)
ax = pylab.gca()
ax.yaxis.set_visible(False)

plotting(hum_10,2,vmin,vmax,cmap,10)
ax = pylab.gca()
ax.yaxis.set_visible(False)

plotting(hum_15,3,vmin,vmax,cmap,15)
ax = pylab.gca()
ax.yaxis.set_visible(False)

plotting(hum_20,4,vmin,vmax,cmap,20)
ax = pylab.gca()
ax.yaxis.set_visible(False)

plotting(hum_25,5,vmin,vmax,cmap,25)
ax = pylab.gca()
ax.yaxis.set_visible(False)

plotting(hum_30,6,vmin,vmax,cmap,30)
ax = pylab.gca()
ax.yaxis.set_visible(False)

plotting(hum_35,7,vmin,vmax,cmap,35)
ax = pylab.gca()
ax.yaxis.set_visible(False)

plotting(hum_40,8,vmin,vmax,cmap,40)
ax = pylab.gca()
ax.yaxis.set_visible(False)

plotting(hum_45,9,vmin,vmax,cmap,45)
ax = pylab.gca()
ax.yaxis.set_visible(False)

plotting(hum_50,10,vmin,vmax,cmap,50)
ax = pylab.gca()
ax.yaxis.set_visible(False)

plotting(hum_60,11,vmin,vmax,cmap,60)
ax = pylab.gca()
ax.yaxis.set_visible(False)

plotting(hum_75,12,vmin,vmax,cmap,75)
ax = pylab.gca()
ax.yaxis.set_visible(False)

mappable = plotting(hum_90,13,vmin,vmax,cmap,90)
ax = pylab.gca()
ax.yaxis.set_visible(False)

ax = pylab.gca()
ax.yaxis.set_visible(False)
cbar = plt.colorbar(mappable, ax=ax)
cbar.ax.set_ylabel('Percentage RH')
plt.suptitle('Relative Humidity with Relation to Control', y=0.9)
plt.show()
```
