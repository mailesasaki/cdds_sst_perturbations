```python
# Load libraries

import xarray as xr
import matplotlib.pyplot as plt
```

```python
# Testing input of north pole perturbation

test_northpole = xr.open_dataset('/data/keeling/a/mailes2/a/CESM_DATA/aqua_northpoleperturb/outputdata/atm/hist/aqua_northpoleperturb.cam.h0.0001-01.nc')
test_control = xr.open_dataset('/data/keeling/a/pappup2/a/CESM/cam5_new/CESM_output_data/atm/hist/cam5_new.cam.h0.0020-12.nc')
test_5 = xr.open_dataset('/data/keeling/a/pappup2/a/CESM/cam_5_n/CESM_output_data/atm/hist/cam_5_n.cam.h0.0020-12.nc')
test_10 = xr.open_dataset('/data/keeling/a/pappup2/a/CESM/cam_10_n/CESM_output_data/atm/hist/cam_10_n.cam.h0.0020-12.nc')
```

# North Pole Perturbations

## Testing Input

```python
test_northpole.mean('lon').TS.plot(label='North Pole Perturbation')
test_control.mean('lon').TS.plot(label='Control')
plt.title('Sea Surface Temperature Test')
plt.legend()
```

```python
test_deltaNP = test_northpole.mean('lon').TS - test_control.mean('lon').TS
test_deltaNP.plot()
plt.title('Delta SST for the North Pole')
```

# 5 Degree Perturbation

```python
test_5.mean('lon').TS.plot(label='5 degree Perturbation')
test_control.mean('lon').TS.plot(label='Control')
plt.title('5 Degree Perturbation')
plt.legend()
```

```python
test_5_delta = test_5.mean('lon').TS - test_control.mean('lon').TS
test_5_delta.plot()
plt.title('Delta SST - 5 degree perturbation')
```

```python
test_10.mean('lon').TS.plot(label='10 Degree Perturbation')
test_control.mean('lon').TS.plot(label='Control')
plt.title('10 Degree Perturbation')
plt.legend()
```
