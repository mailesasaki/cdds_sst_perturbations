# Radiation Response

```python
control_TOA = data_control.FSNT.mean('lon') - data_control.FLNT.mean('lon')
control_TOA.load()

TOA_00 = data_00.FSNT.mean('lon') - data_00.FLNT.mean('lon') - control_TOA
TOA_00.load()

TOA_15 = data_15_n.FSNT.mean('lon') - data_15_n.FLNT.mean('lon') - control_TOA
TOA_15.load()

TOA_45 = data_45_n.FSNT.mean('lon') - data_45_n.FLNT.mean('lon') - control_TOA
TOA_45.load()

TOA_05 = data_05_n.FSNT.mean('lon') - data_05_n.FLNT.mean('lon') - control_TOA
TOA_05.load()

# Surface flux
control_Qs = -data_control.FLNS.mean('lon') + data_control.FSNS.mean('lon') - data_control.SHFLX.mean('lon') - data_control.LHFLX.mean('lon')
control_Qs.load()

Qs_00 = -data_00.FLNS.mean('lon') + data_00.FSNS.mean('lon') - data_00.SHFLX.mean('lon') - data_00.LHFLX.mean('lon') - control_Qs
Qs_00.load()

Qs_15 = -data_15_n.FLNS.mean('lon') + data_15_n.FSNS.mean('lon') - data_15_n.SHFLX.mean('lon') - data_15_n.LHFLX.mean('lon') - control_Qs
Qs_15.load()

Qs_45 = -data_45_n.FLNS.mean('lon') + data_45_n.FSNS.mean('lon') - data_45_n.SHFLX.mean('lon') - data_45_n.LHFLX.mean('lon') - control_Qs
Qs_45.load()

Qs_05 = -data_05_n.FLNS.mean('lon') + data_05_n.FSNS.mean('lon') - data_05_n.SHFLX.mean('lon') - data_05_n.LHFLX.mean('lon') - control_Qs
Qs_05.load()

Qs_10 = -data_10_n.FLNS.mean('lon') + data_10_n.FSNS.mean('lon') - data_10_n.SHFLX.mean('lon') - data_10_n.LHFLX.mean('lon') - control_Qs
Qs_10.load()

#Horizontal Flux
Qa_00 = Qs_00 - TOA_00

Qa_15 = Qs_15 - TOA_15

Qa_45 = Qs_45 - TOA_45
```

```python
plt.rcParams['figure.figsize'] = [15,5]

fig, (ax1, ax2, ax3) = plt.subplots(1,3)


ax1.plot(data_00.coords['lat'], TOA_00, color='blue')
ax1.plot(data_00.coords['lat'], Qs_00, color='red')
ax1.plot(data_00.coords['lat'], Qa_00, color='green')
ax1.set_title('00', fontweight='bold')
ax1.set_ylim(-85,65)
ax1.hlines(y=-85,xmin=-10,xmax=10,colors='black',lw=5)
ax1.yaxis.grid(True, which='major')
ax1.xaxis.grid(True, which='major')
ax1.set_ylabel(r'Radiative Flux ($W/m^2$)')

ax2.plot(data_15_n.coords['lat'], TOA_15, color='blue')
ax2.plot(data_15_n.coords['lat'], Qs_15, color='red')
ax2.plot(data_15_n.coords['lat'], Qa_15, color='green')
ax2.set_title('15', fontweight='bold')
ax2.set_xlabel('Latitude')
ax2.set_ylim(-85,65)
ax2.hlines(y=-85,xmin=5,xmax=25,colors='black',lw=5)
ax2.yaxis.grid(True, which='major')
ax2.xaxis.grid(True, which='major')
for tick in ax2.yaxis.get_major_ticks():
    tick.tick1line.set_visible(False)
    tick.tick2line.set_visible(False)
    tick.label1.set_visible(False)
    tick.label2.set_visible(False)

ax3.plot(data_45_n.coords['lat'], TOA_45, color='blue', label='TOA Flux')
ax3.plot(data_45_n.coords['lat'], Qs_45, color='red', label='Surface Flux')
ax3.plot(data_45_n.coords['lat'], Qa_45, color='green', label='Horiz. Flux')
ax3.set_title('45', fontweight='bold')
ax3.set_ylim(-85,65)
ax3.legend()
ax3.hlines(y=-85,xmin=35,xmax=55,colors='black',lw=5)
ax3.yaxis.grid(True, which='major')
ax3.xaxis.grid(True, which='major')
for tick in ax3.yaxis.get_major_ticks():
    tick.tick1line.set_visible(False)
    tick.tick2line.set_visible(False)
    tick.label1.set_visible(False)
    tick.label2.set_visible(False)

plt.subplots_adjust(wspace=0, hspace=0.0)
plt.show()
```
