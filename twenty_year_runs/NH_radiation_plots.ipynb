# Radiation Fluxes

```python
# TOA Fluxes

TOA_control = data_control.FSNT.mean('lon')-data_control.FLNT.mean('lon')

TOA_pole_n = data_pole_n.FSNT.mean('lon') - data_pole_n.FLNT.mean('lon') - TOA_control
TOA_00 = data_00.FSNT.mean('lon') - data_00.FLNT.mean('lon') - TOA_control
TOA_15_n = data_15_n.FSNT.mean('lon') - data_15_n.FLNT.mean('lon') - TOA_control
TOA_30_n = data_30_n.FSNT.mean('lon') - data_30_n.FLNT.mean('lon') - TOA_control
TOA_45_n = data_45_n.FSNT.mean('lon') - data_45_n.FLNT.mean('lon') - TOA_control
TOA_60_n = data_60_n.FSNT.mean('lon') - data_60_n.FLNT.mean('lon') - TOA_control
TOA_75_n = data_75_n.FSNT.mean('lon') - data_75_n.FLNT.mean('lon') - TOA_control
```

```python
# Surface Fluxes

Qs_control = -data_control.FLNS.mean('lon') + data_control.FSNS.mean('lon') - data_control.SHFLX.mean('lon') - data_control.LHFLX.mean('lon')

Qs_pole_n = -data_pole_n.FLNS.mean('lon') + data_pole_n.FSNS.mean('lon') - data_pole_n.SHFLX.mean('lon') - data_pole_n.LHFLX.mean('lon') - Qs_control
Qs_00 = -data_00.FLNS.mean('lon') + data_00.FSNS.mean('lon') - data_00.SHFLX.mean('lon') - data_00.LHFLX.mean('lon') - Qs_control
Qs_15_n = -data_15_n.FLNS.mean('lon') + data_15_n.FSNS.mean('lon') - data_15_n.SHFLX.mean('lon') - data_15_n.LHFLX.mean('lon') - Qs_control
Qs_30_n = -data_30_n.FLNS.mean('lon') + data_30_n.FSNS.mean('lon') - data_30_n.SHFLX.mean('lon') - data_30_n.LHFLX.mean('lon') - Qs_control
Qs_45_n = -data_45_n.FLNS.mean('lon') + data_45_n.FSNS.mean('lon') - data_45_n.SHFLX.mean('lon') - data_45_n.LHFLX.mean('lon') - Qs_control
Qs_60_n = -data_60_n.FLNS.mean('lon') + data_60_n.FSNS.mean('lon') - data_60_n.SHFLX.mean('lon') - data_60_n.LHFLX.mean('lon') - Qs_control
Qs_75_n = -data_75_n.FLNS.mean('lon') + data_75_n.FSNS.mean('lon') - data_75_n.SHFLX.mean('lon') - data_75_n.LHFLX.mean('lon') - Qs_control
```

```python
# Fickian Diffusion - Horizontal Divergence

Qa_pole_n = Qs_pole_n - TOA_pole_n
Qa_00 = Qs_00 - TOA_00
Qa_15_n = Qs_15_n - TOA_15_n
Qa_30_n = Qs_30_n - TOA_30_n
Qa_45_n = Qs_45_n - TOA_45_n
Qa_60_n = Qs_60_n - TOA_60_n
Qa_75_n = Qs_75_n - TOA_75_n
```
