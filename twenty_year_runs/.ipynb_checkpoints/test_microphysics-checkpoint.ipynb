```python
# Loading libraries

import xarray as xr
import matplotlib.pyplot as plt
import glob
import numpy as np
from matplotlib import gridspec
from mpl_toolkits.axes_grid1 import ImageGrid
import pylab
import matplotlib
```

# Loading Datasets

```python
#North pole perturbation - with MG 1.5 microphysics scheme
data_pole_path = glob.glob("/data/keeling/a/pappup2/a/CESM/cam_90_n/CESM_output_data/atm/hist/*")
data_pole_last18=data_pole_path[24:240] # Get years 3-20#
data_pole_data = xr.open_mfdataset(data_pole_last18,combine="by_coords",use_cftime=True)
data_pole_15 = data_pole_data.mean('time') # Average over the last 18 years
data_pole_15
```

```python
#North pole perturbation - with MG 1.0 microphysics scheme
data_pole_path = glob.glob("/data/keeling/a/mailes2/a/CESM_DATA/aqua_northpoleperturb/outputdata/atm/hist/*")
data_pole_last18=data_pole_path[24:240] # Get years 3-20#
data_pole_data = xr.open_mfdataset(data_pole_last18,combine="by_coords",use_cftime=True)
data_pole_10 = data_pole_data.mean('time') # Average over the last 18 years
data_pole_10
```

# Checking for same SST

```python
data_pole_15.TS.mean('lon').plot(label='MG1.5')
data_pole_10.TS.mean('lon').plot(label='MG1.0')
plt.title('Sea Surface Temperature - Confirmation they are the same')
plt.legend()
```

# Air Temperature

```python
plt.rcParams['figure.figsize'] = [12, 7]

plt.suptitle('Temperature Comparison')
plt.subplot(1,2,1)
data_pole_15.T.mean('lon').plot(yincrease=False)
plt.title('MG 1.5')
plt.subplot(1,2,2)
data_pole_10.T.mean('lon').plot(yincrease=False)
plt.title('MG 1.0')
```

```python
delta_T = data_pole_15.T.mean('lon') - data_pole_10.T.mean('lon')
delta_T.plot(yincrease=False)
plt.title('MG 1.5 - MG 1.0')
```

# TOA Radiation

```python
toa_15 = data_pole_15.FSNT.mean('lon') - data_pole_15.FLNT.mean('lon')
toa_15.load()
toa_10 = data_pole_10.FSNT.mean('lon') - data_pole_10.FLNT.mean('lon')
toa_10.load()
```

```python
plt.suptitle('TOA Comparison')
plt.subplot(1,2,1)
data_pole_15.FLNT.mean('lon').plot(label='LW')
data_pole_15.FSNT.mean('lon').plot(label='SW')
toa_15.plot(yincrease=False, label='TOT')
plt.title('MG 1.5')
plt.subplot(1,2,2)
data_pole_10.FLNT.mean('lon').plot(label='LW')
data_pole_10.FSNT.mean('lon').plot(label='SW')
toa_10.plot(yincrease=False, label='TOT')
plt.title('MG 1.0')
plt.legend()
```

```python
delta_FLNT = data_pole_15.FLNT.mean('lon') - data_pole_10.FLNT.mean('lon')
delta_FLNT.load()
delta_FSNT = data_pole_15.FSNT.mean('lon') - data_pole_10.FSNT.mean('lon')
delta_FSNT.load()
delta_toa = toa_15 - toa_10
```

```python
delta_FLNT.plot(label='FLNT')
delta_FSNT.plot(label='FSNT')
delta_toa.plot(label='TOT')
plt.title('MG 1.5 - MG 1.0')
plt.ylabel('Radiation (W/m2)')
plt.legend()
```

# Radiation - Qs and Qa as well

```python
Qs_15 = -data_pole_15.FLNS.mean('lon') + data_pole_15.FSNS.mean('lon') - data_pole_15.SHFLX.mean('lon') - data_pole_15.LHFLX.mean('lon')
Qs_15.load()

Qs_10 = -data_pole_10.FLNS.mean('lon') + data_pole_10.FSNS.mean('lon') - data_pole_10.SHFLX.mean('lon') - data_pole_10.LHFLX.mean('lon')
Qs_10.load()
```

```python
Qa_15 = Qs_15 - toa_15
Qa_10 = Qs_10 - toa_10
```

```python
plt.suptitle('TOA Comparison')
plt.subplot(1,2,1)
toa_15.plot(label='TOA', color='blue')
Qs_15.plot(label='Qs', color='red')
Qa_15.plot(label='TOT', color='green')
plt.ylabel('Radiation (W/m2)')
plt.title('MG 1.5')
plt.subplot(1,2,2)
toa_10.plot(label='TOA', color='blue')
Qs_10.plot(label='Qs', color='red')
Qa_10.plot(label='Qa', color='green')
plt.title('MG 1.0')
plt.legend()
```

```python
delta_qs = Qs_15 - Qs_10
delta_qa = Qa_15 - Qa_10
```

```python
delta_qs.plot(label='Qs',color='red')
delta_qa.plot(label='Qa',color='green')
delta_toa.plot(label='TOA',color='blue')
plt.title('MG1.5 - MG1.0 - Radiation Values')
plt.ylabel('Radiation (W/m2)')
plt.legend()
```
