# SST 

```python
sst_control = data_control.TS.mean('lon')
sst_control.load()

sst_00 = data_00.TS.mean('lon') - sst_control
sst_00.load()
sst_15 = data_15.TS.mean('lon') - sst_control
sst_15.load()
sst_30 = data_30.TS.mean('lon') - sst_control
sst_30.load()
sst_45 = data_extratropic.TS.mean('lon') - sst_control
sst_45.load()
sst_60 = data_60.TS.mean('lon') - sst_control
sst_60.load()
sst_75 = data_75.TS.mean('lon') - sst_control
sst_75.load()
```

# TOA Flux

```python
toa_control = data_control.FSNT.mean('lon')-data_control.FLNT.mean('lon')
toa_control.load()

#TOA_pole = data_pole.FSNT.mean('lon') - data_pole.FLNT.mean('lon') - toa_control
TOA_15 = data_15.FSNT.mean('lon') - data_15.FLNT.mean('lon') - toa_control
TOA_15.load()
TOA_30 = data_30.FSNT.mean('lon') - data_30.FLNT.mean('lon') - toa_control
TOA_30.load()
TOA_45 = data_extratropic.FSNT.mean('lon') - data_extratropic.FLNT.mean('lon') - toa_control
TOA_45.load()
TOA_00 = data_00.FSNT.mean('lon') - data_00.FLNT.mean('lon') - toa_control
TOA_00.load()
TOA_60 = data_60.FSNT.mean('lon') - data_60.FLNT.mean('lon') - toa_control
TOA_60.load()
TOA_75 = data_75.FSNT.mean('lon') - data_75.FLNT.mean('lon') - toa_control
TOA_75.load()
```

# Surface Flux

```python
#Surface flux
Qs_control = -data_control.FLNS.mean('lon') + data_control.FSNS.mean('lon') - data_control.SHFLX.mean('lon') - data_control.LHFLX.mean('lon')
Qs_control.load()

#Qs_pole = -data_pole.FLNS.mean('lon') + data_pole.FSNS.mean('lon') - data_pole.SHFLX.mean('lon') - data_pole.LHFLX.mean('lon') - Qs_control
#Qs_pole.load()
Qs_00 = -data_00.FLNS.mean('lon') + data_00.FSNS.mean('lon') - data_00.SHFLX.mean('lon') - data_00.LHFLX.mean('lon') - Qs_control
Qs_00.load()
Qs_15 = -data_15.FLNS.mean('lon') + data_15.FSNS.mean('lon') - data_15.SHFLX.mean('lon') - data_15.LHFLX.mean('lon') - Qs_control
Qs_15.load()
Qs_30 = -data_30.FLNS.mean('lon') + data_30.FSNS.mean('lon') - data_30.SHFLX.mean('lon') - data_30.LHFLX.mean('lon') - Qs_control
Qs_30.load()
Qs_45 = -data_extratropic.FLNS.mean('lon') + data_extratropic.FSNS.mean('lon') - data_extratropic.SHFLX.mean('lon') - data_extratropic.LHFLX.mean('lon') - Qs_control
Qs_45.load()
Qs_60 = -data_60.FLNS.mean('lon') + data_60.FSNS.mean('lon') - data_60.SHFLX.mean('lon') - data_60.LHFLX.mean('lon') - Qs_control
Qs_60.load()
Qs_75 = -data_75.FLNS.mean('lon') + data_75.FSNS.mean('lon') - data_75.SHFLX.mean('lon') - data_75.LHFLX.mean('lon') - Qs_control
Qs_75.load()
```
