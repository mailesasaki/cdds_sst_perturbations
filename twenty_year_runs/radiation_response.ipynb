# Take global means over the radiation response

```python
# Finding TOA radiation

#Taking global mean shortwave flux and subtracting global mean longwave flux - is this correct?

control_R = data_control.FSNT.mean(('lat','lon')) - data_control.FLNT.mean(('lat','lon')) #Top of atmosphere radiation - incoming SW - outgoing LW

data_00_R = data_00.FSNT.mean(('lat','lon')) - data_00.FLNT.mean(('lat','lon'))
data_15_R = data_15_n.FSNT.mean(('lat','lon')) - data_15_n.FLNT.mean(('lat','lon'))
data_30_R = data_30_n.FSNT.mean(('lat','lon')) - data_30_n.FLNT.mean(('lat','lon'))
data_45_R = data_45_n.FSNT.mean(('lat','lon')) - data_45_n.FLNT.mean(('lat','lon'))
data_60_R = data_60_n.FSNT.mean(('lat','lon')) - data_60_n.FLNT.mean(('lat','lon'))
data_75_R = data_75_n.FSNT.mean(('lat','lon')) - data_75_n.FLNT.mean(('lat','lon'))
data_pole_R = data_pole_n.FSNT.mean(('lat','lon')) - data_pole_n.FLNT.mean(('lat','lon'))
```

```python
R = [data_00_R,data_15_R,data_30_R,data_45_R,data_60_R,data_75_R,data_pole_R]
lats = [0,15,30,45,60,75,90]

plt.scatter(lats,R)
plt.title('Radiation by latitude of perturbation')
plt.ylabel('Radiation (W/m2)')
plt.xlabel('Latitude')
```

```python
#Difference in TOA radiation response

data_00_dR = data_00_R - control_R
data_15_dR = data_15_R - control_R
data_30_dR = data_30_R - control_R
data_45_dR = data_45_R - control_R
data_60_dR = data_60_R - control_R
data_75_dR = data_75_R - control_R
data_pole_dR = data_pole_R - control_R

dR = [data_00_dR,data_15_dR,data_30_dR,data_45_dR,data_60_dR,data_75_dR,data_pole_dR]
```

```python
#Plot TOA response

lats = [0,15,30,45,60,75,90]

plt.scatter(lats,dR)
plt.title('Difference in Radiation from Control')
plt.ylabel('Radiation (W/m2)')
plt.xlabel('Latitude')
```
