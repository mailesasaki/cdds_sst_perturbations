```python
# Loading libraries

import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Range of latitudes
lats = np.arange(-np.pi/2, np.pi/2, np.pi/180)
print(lats)
print(len(lats))
```

```python
# Setting a base SST
SST = np.zeros(180)
i = 0
for lat in lats:
    if lat < -np.pi/3:
        SST[i] = 0
    elif lat > np.pi/3:
        SST[i] = 0
    else:
        SST[i] = 27 * (1 - np.sin(3*lat/2)**2)
    i += 1
```

```python
#Plotting Control SST
lats_degrees = lats * 180/np.pi
plt.plot(lats_degrees, SST)
plt.title("Control SST")
plt.xlabel("Latitude")
plt.ylabel("SST (C)")
```

```python
max_lat = lats[89]

amplitude = 2
width = 50*np.pi/180
SST_exp = list(SST) # Not the same array

j = 0
for lat in lats:
    if (lat < max_lat + width) and (lat > max_lat - width):
        SST_exp[j] += amplitude * np.cos((np.pi/2) * ((lat-max_lat)/width))**2
    j += 1
SST_exp_00 = np.array(SST_exp)
```

```python
#Plotting Equator
plt.rcParams['figure.figsize'] = [10,5]

lats_degrees = lats * 180/np.pi
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(lats_degrees, SST_exp_00, label='00 patch')
ax1.plot(lats_degrees, SST, label='Control')
ax1.set_xlabel("Latitude")
ax1.set_ylabel("SST (C)")
ax1.set_title('Total')
ax1.legend()

ax2.plot(lats_degrees, SST_exp_00-SST)
ax2.set_xlabel("Latitude")
ax2.set_title('Perturbation')

plt.suptitle("Equator Perturb")
```

# Adding all SSTs up

```python
SST_tot = SST_exp_00 + SST_exp_25 + SST_exp_50 + SST_exp_75 + SST_exp_90 + SST_exp_125 + \
            SST_exp_25_s + SST_exp_50_s + SST_exp_75_s + SST_exp_90_s + SST_exp_125_s
SST_list = [SST_exp_00, SST_exp_25, SST_exp_50, SST_exp_75, SST_exp_90, SST_exp_125, SST_exp_25_s, SST_exp_50_s, SST_exp_75_s,
            SST_exp_90_s, SST_exp_125_s]
patches = [00,25,50,75,100,125,-25,-50,-75,-100,-125]
```

```python
plt.rcParams['figure.figsize'] = [10,5]

lats_degrees = lats * 180/np.pi
fig, ax1 = plt.subplots(1,1)

ax1.plot(lats_degrees, SST_tot-11*SST, label='All added together')
print(SST_tot-11*SST)
ax1.set_xlabel("Latitude")
ax1.set_ylabel("SST (C)")
ax1.set_title('Total')
ax1.set_ylim(2,6)
ax1.legend()

plt.suptitle("Total Perturbation")
```

```python
plt.rcParams['figure.figsize'] = [12,7]

lats_degrees = lats * 180/np.pi
fig, (ax1,ax2) = plt.subplots(1,2)

for patch in SST_list:
    ax1.plot(lats_degrees, patch)
ax1.set_title('Total')    
ax1.set_xlabel('Latitude')
ax1.set_ylabel('SST (C)')

i = 0
for patch in SST_list:
    ax2.plot(lats_degrees, patch-SST, label=patches[i])
    i += 1
ax2.legend()
ax2.set_title('Perturbations')
ax2.set_xlabel('Latitude')
```
