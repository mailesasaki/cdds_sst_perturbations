```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Range of latitudes
lats = np.arange(-np.pi/2, np.pi/2, np.pi/180)
print(lats)
print(len(lats))
```

# Control SST

```python
# Setting a base SST
SST = np.zeros(180)
i = 0
for lat in lats:
    if lat < -np.pi/3:
        SST[i] = 0
    elif lat > np.pi/3:
        SST[i] = 0
    else:
        SST[i] = 27 * (1 - np.sin(3*lat/2)**2)
    i += 1
```

```python
#Plotting Control SST
lats_degrees = lats * 180/np.pi
plt.plot(lats_degrees, SST)
plt.title("Control SST")
plt.xlabel("Latitude")
plt.ylabel("SST (C)")
```

# Symmetric Perturbations

## Equator Perturbation

```python
# Adding perturbation at the equator

max_lat = lats[90]
amplitude = 2
width = 10*np.pi/180
SST_exp = list(SST) # Not the same array
j = 0
for lat in lats:
    if (lat < max_lat + width) and (lat > max_lat - width):
        SST_exp[j] += amplitude * np.cos((np.pi/2) * ((lat-max_lat)/width))**2
    j += 1
SST_exp = np.array(SST_exp)
delta_SST_0 = SST_exp - SST
```

```python
# Plotting perturbation at the equator
plt.plot(lats_degrees, SST_exp)
plt.plot(lats_degrees, SST)
plt.title("Equator patch")
plt.xlabel("Latitude")
plt.ylabel("SST (C)")
```

```python
# Delta SST
plt.plot(lats_degrees, delta_SST_0)
plt.title('Equator Perturbation')
plt.xlabel('Latitude')
plt.ylabel('SST (C)')
```
